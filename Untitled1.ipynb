```python
airbnb_reviews=airbnb_df.groupby('neighbourhood_group')['number_of_reviews'].count().reset_index()
airbnb_reviews
```

```python
airbnb_reviews=airbnb_df.groupby('neighbourhood_group')['number_of_reviews'].min().reset_index()
airbnb_reviews
```

```python
airbnb_reviews=airbnb_df.groupby('room_type')['number_of_reviews'].mean().reset_index()
airbnb_reviews
```

Reviews given to room_type shared room are less compared to room type private room and entire home/apt. Hence consider to improve the quality of service by solving customer queries.




```python
airbnb_reviews=airbnb_df.groupby('minimum_nights')['number_of_reviews'].count().reset_index()
airbnb_reviews
```

As we can see from above code that reviews from customers who have booked for minimum nights are higher compared to customers who booked hotel for more than 480 days to stay.
Hence the customer with maximum stay should be of top priority for business.
Get to know the concerns and trouble faced by such customers and take  good reviews from them.
