```python
#Seeing which neighborhood has the highest number of listings
feq = airbnb_df['neighbourhood_group'].value_counts().sort_values(ascending=True)

feq.plot.barh(figsize=(8,4), color ='b', width= 0.5)
plt.title('Number of listings by neighbourhood', fontsize=14)
plt.xlabel('calculated_host_listings_count', fontsize = 12)
plt.show()
```

```python
airbnb_df.head()
plt.figure(figsize=(12,8))
df = airbnb_df[airbnb_df['minimum_nights']==1 ]
df1 = df.groupby(['room_type','neighbourhood_group'])['price'].mean().sort_values(ascending=True)
df1.plot(kind='bar')
plt.title('Average Price for rooms in neighbourhood group')
plt.ylabel('Average Daily Price')
plt.xlabel('Neighbourhood Group')
plt.show()
print('List of Average Price per night based on the neighbourhood group')
pd.DataFrame(df1).sort_values(by='room_type')
```

The above list gives per day charges required for airbnb bookings around neighbourhood_group.

```python
print('Top 20 most expensive locality in Airbnb listing are :')
df_expensive = airbnb_df.dropna(subset=["price"]).groupby("neighbourhood")[["neighbourhood", "price"]].agg("mean").sort_values(by="price",ascending=False).head(15)
df_expensive.plot(kind='bar')
plt.show()
pd.DataFrame(df_expensive)
```

Fort Wadsworth is the most expensive among neighbourhood with price 800.
